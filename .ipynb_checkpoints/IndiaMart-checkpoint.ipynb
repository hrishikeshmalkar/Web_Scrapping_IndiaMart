#### Main code : Fetching Data as per requirements and store it into lists.

```python
# Lists to store the scraped data in
Names = []
Mobile = []
Firm = []
Address = []
Email = []
Phone = []
Fax = []
Route = []
B_Date = []
J_Date = []


# Extract data from individual Data container
for container in data:
# The name
    name = container.h3.text
    Names.append(name)
    
# The Mobile Number
    mobile = container.p.text
    Mobile.append(mobile)
    
    
    
    dicts = {}

    i1=[]
    i2=[]

    for item in container.find_all('div', class_ = 'b-branches__item'):
        i1.append(item.b.text)
    for item in container.find_all('div', class_ = 'b-branches__item'):
        i2.append(item.text)
    
    keys = i1
    values = i2
    
    
    res = {keys[i]: values[i] for i in range(len(keys))}
    
    
    
    val='NULL'
    val1='NULL'
    val2='NULL'
    val3='NULL'
    val4='NULL'
    val5='NULL'
    val6='NULL'
    val7='NULL'

    for i in res.keys():

        if i == 'Firm':
            val=res['Firm'] 

        if i=='Address':
            val1=res['Address']

        if i =='Email':
            val2=res['Email']

        if i == 'Phone No':
            val3=res['Phone No']

        if i=='Fax No':
            val4=res['Fax No']

        if i =='Route':
            val5=res['Route']

        if i == 'B. Date':
            val6=res['B. Date']

        if i == 'J. Date':
            val7=res['J. Date']

    Firm.append(val)
    Address.append(val1)
    Email.append(val2)
    Phone.append(val3)
    Fax.append(val4)
    Route.append(val5)
    B_Date.append(val6)
    J_Date.append(val7)
```

#### Creating DataFrame

```python
df = pd.DataFrame(list(zip(Names, Mobile, Firm, Address, Email, Phone, Fax, Route, B_Date, J_Date)), 
               columns =['Names', 'Mobile', 'Firm', 'Address', 'Email', 'Phone', 'Fax', 'Route', 'B_Date', 'J_Date']) 
df.head()
```

```python
df.index += 1
```

#### Preview of dataset

```python
df.head()
```

```python
df.tail()
```

#### Data Pre-processing 1

```python
df['Mobile']=df['Mobile'].str.replace('Mobile No ','')
df['Mobile']=df['Mobile'].str.replace(' ','')
df['Mobile']=df['Mobile'].str.replace('?','')
df['Firm']=df['Firm'].str.replace('Firm ','')
df['Address']=df['Address'].str.replace('Address ','')
df['Email']=df['Email'].str.replace('Email ', '')
df['Phone']=df['Phone'].str.replace('Phone No ','')
df['Fax']=df['Fax'].str.replace('Fax No ','')
df['Route']=df['Route'].str.replace('Route ','')
df['B_Date']=df['B_Date'].str.replace('B. Date ','')
df['J_Date']=df['J_Date'].str.replace('J. Date','')
df['J_Date']=df['J_Date'].str.replace('LIFETIME MEMBER','')
```
